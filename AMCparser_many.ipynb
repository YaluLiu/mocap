```python
def make_dirs(human_idx,action_idx):
    dir_path = "./output_human/" + str(human_idx).zfill(2) + "/" + str(action_idx).zfill(2)
    if not os.path.isdir(dir_path):
        os.makedirs(dir_path)
    for cam_idx in range(6):
        cam_path = dir_path + "/cam" + str(cam_idx)
        if not os.path.isdir(cam_path):
            os.makedirs(cam_path)
    return dir_path
```

```python
def make_points(human_idx,action_idx):
    asf_path = './data/asf/' + str(human_idx).zfill(2) + ".asf"
    amc_path = './data/amc/' + str(human_idx).zfill(2) + '_' + str(action_idx).zfill(2) + ".amc"
    print(asf_path, amc_path)
    if not os.path.isfile(asf_path):
        return
    if not os.path.isfile(amc_path):
        return
    amcparser = AMCParser(asf_path,amc_path)
    amcparser.parse()
    #trans asf points to human3.6M 17 points
    human_nodes_lst = trans_joints_lst(amcparser.joints_lst,"asf")
    dir_path = make_dirs(human_idx,action_idx)
    #output skeleton to json and points to npy
    json_path = dir_path + '/points_3d.json'
    npy_path  = dir_path + '/points_3d.npy'
    #save 3d points data and special solve
    #because person is too huge, move foot to origin point
    save_data(human_nodes_lst,json_path,npy_path,dimension=3,special_solve=True)
```

```python
def parse_by_num(end_person = 5,end_action = 5):
    for person_idx in range(7,end_person):
        for action_idx in range(1,end_action):
            make_points(person_idx,action_idx)
    print("Work Done!")
```
